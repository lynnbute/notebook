## Converting date to independent columns

```python
#convert the datetime column to actual datetime
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["Year"] = df["Date"].dt.year
```

```python
df["Month"] = df["Date"].dt.month
df["Days"] = df["Date"].dt.day
```

```python
df.head()
```

```python
df = df[df.columns[1:]]
df.head()
```

```python
lynatte.nunique()
```

```python
lynatte['Open'].unique()
```

```python
df.isnull().sum()
```

```python
yData = df["Volume"]
```

```python
XData = df[[column for column in df if (column != "Volume")]]
```

```python
XData.head()
```

## SPLITTING DATA INTO TRAINING ,VALIDATION AND TESTING

```python
from sklearn.model_selection import train_test_split
```

```python
Xtrain,Xtest, ytrain, ytest = train_test_split(XData, yData, test_size =0.15,random_state=42)
```

```python
len(Xtrain)
```

```python
len(Xtest)
```

```python
Xtrain.head()
```

```python
Xtest.head()
```

## Adding validation dataset

```python
#Validation set used to validate the model through training
```

```python
Xtrain,Xval,ytrain,yval = train_test_split(Xtrain, ytrain, test_size =0.15,random_state=42)
```

```python
len(Xtrain)
```

```python
len(Xval)
```

## Model fitting


```python
# Model fitting is a measure of how well a machine learning model generalizes to similar data to that on which it was trained.
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor()
result = model.fit(Xtrain, ytrain)
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, accuracy_score
```

```python
predicted_test = model.predict(Xtest)
```

```python
model.score(Xtrain, ytrain)
```

```python
model.score(Xtest, ytest)
```

## VALIDATION SET
